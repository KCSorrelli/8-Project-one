```python
year_df = pd.DataFrame()
decade_df = pd.DataFrame()
for year in range(1960,2010):
    time.sleep(0.25)
    response = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=' 
                        +  api_key + '&primary_release_year=' + str(year) + '&sort_by=revenue.desc&language=en-US')
                           
    #store parsed json response
    highest_revenue = response.json() 
    
    # highest_revenue
    highest_revenue_films = highest_revenue['results']

    data_df = pd.DataFrame(highest_revenue_films)
    
    data_df['release_date'] = pd.to_datetime(data_df['release_date'])

    year_df[str(year)] = data_df['title'] 
    
    for film in highest_revenue_films:
        film_revenue = requests.get('https://api.themoviedb.org/3/movie/'+ 
                                    str(film['id']) +'?api_key='+ api_key+'&language=en-US')
        time.sleep(0.25)
        film_revenue = film_revenue.json()
        film_revenue['release_date'] = pd.to_datetime(film_revenue['release_date'])
        film_revenue['year'] = film_revenue['release_date'].year
        print(film_revenue)
        mylist = [[data_df], [film_revenue]]
        ict_df = pd.DataFrame([ pd.Series(value) for value in mylist ])
        #ict_df.head()
        print(locale.currency(film_revenue['revenue'], grouping=True ))
        # store title and revenue in our dataframe 
        df.loc[len(df)]=[film['title'],film_revenue['revenue'],film['release_date']] 

```
